```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("car price.csv")
df
```

missing value

```python
df.isna().sum()
```

```python
mean_kil = df['Kilometers_Driven'].mean()
```

```python
df['Kilometers_Driven'] = df['Kilometers_Driven'].fillna(mean_kil,axis=0)
```

```python
mean_Airbags = df['Airbags'].mean()
df['Airbags'] = df['Airbags'].fillna(mean_Airbags,axis=0)
```

```python
df.isna().sum()
```

### outlier

```python
sns.jointplot(data=df , x="Airbags" , y="Price")
```

```python
import numpy as np
```

```python
np.percentile(df['Airbags'],[99])[0]
```

```python
upper_limit = np.percentile(df['Airbags'],[99])[0]
upper_limit
```

```python
df[df.Airbags > upper_limit]
```

```python
df.Airbags[(df.Airbags > 3*upper_limit)] = 3*upper_limit
```
