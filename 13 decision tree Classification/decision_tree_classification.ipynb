```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Movie_classification.csv' , header=0 )
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Time_taken'].mean()
```

```python
# df['Time_taken'].fillna(df['Time_taken'].mean() , inplace=True) # for permanent change
```

```python
df['Time_taken'] = df['Time_taken'].fillna(df['Time_taken'].mean() , axis=0)
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df,columns=['3D_available','Genre'] , drop_first=True) # drop_first convert column to get output as n-1 
```

```python
df.head()
```

```python
X = df.drop('Start_Tech_Oscar',axis=1)
y=df['Start_Tech_Oscar']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,random_state=42 , test_size=0.2)
```

### training_regression tree

```python
from sklearn.tree import  DecisionTreeClassifier
```

```python
help(DecisionTreeClassifier())
```

```python
classtree = DecisionTreeClassifier(max_depth=4)
```

```python
X_train.shape
```

```python
classtree.fit(X_train , y_train)
```

```python
y_train_pred = classtree.predict(X_train)
y_test_pred = classtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , mean_squared_error
```

```python
mean_squared_error(y_test , y_test_pred)
```

```python
confusion_matrix(y_test , y_test_pred)
```

```python
confusion_matrix(y_train , y_train_pred)
```

```python
accuracy_score(y_test , y_test_pred)
```

### plot decision tree

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(14, 8))
plot_tree(classtree, filled=True, feature_names=y)
plt.title("Decision Tree Regressor")
plt.show()
```
